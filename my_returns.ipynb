```python
# @title
# Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title
# load the data
#sp500 = pd.read_csv('SPX.csv')
#vnq = pd.read_csv('VNQ.csv')

```

```python
# @title
NIFTY_50 = pd.read_csv('NIFTY 50.csv')
NIFTY_REALTY = pd.read_csv('NIFTY REALTY.csv')
NIFTY_BANK = pd.read_csv('NIFTY BANK.csv')
NIFTY_NEXT50 = pd.read_csv('NIFTY NEXT 50.csv')
NIFTY_LARGEMIDCAP_250 = pd.read_csv('NIFTY LARGEMIDCAP 250.csv')
NIFTY_PHARMA = pd.read_csv('NIFTY PHARMA.csv')
NIFTY_ENERGY = pd.read_csv('NIFTY ENERGY.csv')
NIFTY_IT = pd.read_csv('NIFTY IT.csv')
#NIFTY_REALTY.info()
```

```python
# @title
# calculate daily returns
col = 'Close'
#sp500['daily_returns'] = sp500[col].pct_change()
#vnq['daily_returns'] = vnq[col].pct_change()
NIFTY_50['daily_returns'] = NIFTY_50[col].pct_change()
NIFTY_REALTY['daily_returns'] = NIFTY_REALTY[col].pct_change()
NIFTY_BANK['daily_returns'] = NIFTY_BANK[col].pct_change()
NIFTY_NEXT50['daily_returns'] = NIFTY_NEXT50[col].pct_change()
NIFTY_LARGEMIDCAP_250['daily_returns'] = NIFTY_LARGEMIDCAP_250[col].pct_change()
NIFTY_PHARMA['daily_returns'] = NIFTY_PHARMA[col].pct_change()
NIFTY_ENERGY['daily_returns'] = NIFTY_ENERGY[col].pct_change()
NIFTY_IT['daily_returns'] = NIFTY_IT[col].pct_change()
```

```python
# @title
# calculate cumulative returns
#sp500['cum_returns'] = (1 + sp500['daily_returns']).cumprod()
#vnq['cum_returns'] = (1 + vnq['daily_returns']).cumprod()
NIFTY_50['cum_returns'] = (1 + NIFTY_50['daily_returns']).cumprod()
NIFTY_REALTY['cum_returns'] = (1 + NIFTY_REALTY['daily_returns']).cumprod()
NIFTY_BANK['cum_returns'] = (1 + NIFTY_BANK['daily_returns']).cumprod()
NIFTY_NEXT50['cum_returns'] = (1 + NIFTY_NEXT50['daily_returns']).cumprod()
NIFTY_LARGEMIDCAP_250['cum_returns'] = (1 + NIFTY_LARGEMIDCAP_250['daily_returns']).cumprod()
NIFTY_PHARMA['cum_returns'] = (1 + NIFTY_PHARMA['daily_returns']).cumprod()
NIFTY_ENERGY['cum_returns'] = (1 + NIFTY_ENERGY['daily_returns']).cumprod()
NIFTY_IT['cum_returns'] = (1 + NIFTY_IT['daily_returns']).cumprod()
```

```python
# @title
# print cumulative returns
#print('sp500: ' + str(sp500['cum_returns'].tail(1)))
#print('vnq: ' +str(vnq['cum_returns'].tail(1)))
print('cumulative returns of NIFTY 50: ' +str(NIFTY_50['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY REALTY: ' +str(NIFTY_REALTY['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY BANK: ' +str(NIFTY_BANK['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY NEXT 50: ' +str(NIFTY_NEXT50['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY LARGEMIDCAP 250: ' +str(NIFTY_LARGEMIDCAP_250['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY PHARMA: ' +str(NIFTY_PHARMA['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY ENERGY: ' +str(NIFTY_ENERGY['cum_returns'].iloc[-1]))
print('cumulative returns of NIFTY IT: ' +str(NIFTY_IT['cum_returns'].iloc[-1]))
```

```python
# @title
# plot the retuns
#sp500['cum_returns'].plot(label ='S&P 500')
#vnq['cum_returns'].plot(label = 'VNQ')
NIFTY_50['cum_returns'].plot(label = 'NIFTY 50')
NIFTY_REALTY['cum_returns'].plot(label = 'NIFTY REALTY')
#NIFTY_PHARMA['cum_returns'].plot(label = 'NIFTY PHARMA')
NIFTY_ENERGY['cum_returns'].plot(label = 'NIFTY ENERGY')
#NIFTY_IT['cum_returns'].plot(label = 'NIFTY IT')
#NIFTY_BANK['cum_returns'].plot(label = 'NIFTY BANK')
NIFTY_NEXT50['cum_returns'].plot(label = 'NIFTY NEXT 50')
NIFTY_LARGEMIDCAP_250['cum_returns'].plot(label='NIFTY LARGEMIDCAP 250')
plt.xlabel('Days')
plt.ylabel('Cumulative Returns')
plt.title('Cumulative Returns of investement in stocks vs Reality')
plt.legend()
```
