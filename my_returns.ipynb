```python
# @title
# Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# @title
# load the data
#sp500 = pd.read_csv('SPX.csv')
#vnq = pd.read_csv('VNQ.csv')

import yfinance as yf

# Download the stock data

#stk="^NSEI"
#tkr = yf.Ticker(stk)
#data=tkr.history(period="max")
#data.to_csv(stk+'.csv')

stk = yf.Tickers('^NSEI ^CNXREALTY ^CNXENERGY ^CNXIT 0P0001RQX5.BO')
for tkr in stk.tickers:
  stkname=stk.tickers[tkr].info['shortName']
  #print(stkname)
  data=stk.tickers[tkr].history(period="1y")
  #print(data.head())
  data.to_csv(tkr+'.csv')
  tkr = pd.read_csv(tkr+'.csv') #/content/sample_data

  col='Close'

  tkr['daily_returns'] = tkr[col].pct_change()
  tkr['cum_returns'] = (1 + tkr['daily_returns']).cumprod()

  #print(tkr['cum_returns'].iloc[-1])

  tkr['cum_returns'].plot(label = stkname)

  #NIFTY_50 = pd.read_csv(tkr+'.csv'
  #NIFTY_REALTY = pd.read_csv(tkr'.csv')
  #tkr = yf.Ticker(ticker)
  #tkr.history(period="max")
  #print(tkr)
#  data=tkr.history(period="max")
#  data.to_csv(stk+'.csv')
plt.xlabel('#Days')
plt.ylabel('Cumulative Returns %')
plt.title('Cumulative Returns of investement in Stocks vs Real-Estate')
plt.legend()
```

```python
# @title

NIFTY_BANK = pd.read_csv('/content/sample_data/NIFTY BANK.csv')
NIFTY_NEXT50 = pd.read_csv('/content/sample_data/NIFTY NEXT 50.csv')
NIFTY_LARGEMIDCAP_250 = pd.read_csv('/content/sample_data/NIFTY LARGEMIDCAP 250.csv')
NIFTY_PHARMA = pd.read_csv('/content/sample_data/NIFTY PHARMA.csv')
NIFTY_ENERGY = pd.read_csv('/content/sample_data/NIFTY ENERGY.csv')
NIFTY_IT = pd.read_csv('/content/sample_data/NIFTY IT.csv')
#NIFTY_REALTY.info()
```

```python
# @title
# calculate daily returns
col = 'Close'
#sp500['daily_returns'] = sp500[col].pct_change()
#vnq['daily_returns'] = vnq[col].pct_change()
NIFTY_50['daily_returns'] = NIFTY_50[col].pct_change()
NIFTY_REALTY['daily_returns'] = NIFTY_REALTY[col].pct_change()
NIFTY_BANK['daily_returns'] = NIFTY_BANK[col].pct_change()
NIFTY_NEXT50['daily_returns'] = NIFTY_NEXT50[col].pct_change()
NIFTY_LARGEMIDCAP_250['daily_returns'] = NIFTY_LARGEMIDCAP_250[col].pct_change()
NIFTY_PHARMA['daily_returns'] = NIFTY_PHARMA[col].pct_change()
NIFTY_ENERGY['daily_returns'] = NIFTY_ENERGY[col].pct_change()
NIFTY_IT['daily_returns'] = NIFTY_IT[col].pct_change()
```
